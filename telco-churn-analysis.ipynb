```python
# Import libraries
import pandas as pd
import numpy as np

# Load dataset
file_path = r"C:\Users\USER\Downloads\WA_Fn-UseC_-Telco-Customer-Churn.csv"
df = pd.read_csv(file_path)

# Display the first 5 rows
df.head()
```

```python
# Shape of the data
print("Shape:", df.shape)

# Columns and data types
print("\nInfo:")
df.info()

# Summary statistics
print("\nSummary Statistics:")
df.describe(include='all')
```

```python
# Convert 'TotalCharges' to numeric (some might be blank)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Drop rows with missing TotalCharges
df.dropna(subset=['TotalCharges'], inplace=True)

# Reset index after drop
df.reset_index(drop=True, inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Churn count plot
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Churn', hue='Churn', palette='Set2', legend=False)
plt.title("Churn Distribution")
plt.show()



# Churn by contract type
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Contract', hue='Churn', palette='Set1')
plt.title("Churn by Contract Type")
plt.show()

# Monthly charges vs. churn
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='MonthlyCharges', hue='Churn', kde=True, palette='coolwarm')
plt.title("Monthly Charges Distribution by Churn")
plt.show()

# Boxplot: tenure vs. churn
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Churn', y='tenure', hue='Churn', palette='Set3', dodge=False, legend=False)
plt.title("Tenure by Churn")
plt.show()
```

```python
# Encode Churn as binary
df['Churn_binary'] = df['Churn'].map({'Yes':1, 'No':0})

# Numerical correlation heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='Blues')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# High risk churn segment
high_churn = df[(df['Churn'] == 'Yes') & (df['Contract'] == 'Month-to-month')]

# Insight 1: What % of churn comes from Month-to-month customers?
percent_monthly = len(high_churn) / len(df[df['Churn'] == 'Yes']) * 100
print(f"🔍 Insight: {percent_monthly:.2f}% of all churned customers were on Month-to-month contracts.")

# Insight 2: Senior citizens more likely to churn?
senior_churn_rate = df[df['SeniorCitizen'] == 1]['Churn_binary'].mean()
non_senior_churn_rate = df[df['SeniorCitizen'] == 0]['Churn_binary'].mean()
print(f"🔍 Insight: Senior Citizen Churn Rate = {senior_churn_rate:.2f}")
print(f"🔍 Insight: Non-Senior Citizen Churn Rate = {non_senior_churn_rate:.2f}")
```
